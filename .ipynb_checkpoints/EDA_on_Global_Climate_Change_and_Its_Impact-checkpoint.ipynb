```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/GlobalLandTemperaturesByMajorCity.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
# Check Data Types
df.info()
```

```python
#Convert dt to Datetime
df['dt'] = pd.to_datetime(df['dt'])
```

```python
# Ensure Latitude and Longitude are treated as strings
df['Latitude'] = df['Latitude'].astype(str)
df['Longitude'] = df['Longitude'].astype(str)

# Now remove the directional suffix and convert Latitude to numerical values
df['Latitude'] = df['Latitude'].str[:-1].astype(float)  # Remove the last character (N/S)
df['Latitude'] = df['Latitude'].where(df['Latitude'] > 0, -df['Latitude'])  # Adjust for S (negative values)

# Remove the directional suffix and convert Longitude to numerical values
df['Longitude'] = df['Longitude'].str[:-1].astype(float)  # Remove the last character (E/W)
df['Longitude'] = df['Longitude'].where(df['Longitude'] > 0, -df['Longitude'])  # Adjust for W (negative values)

# Verify the conversion
print(df[['Latitude', 'Longitude']].head())
print(df[['Latitude', 'Longitude']].tail())
```

```python
# Handle the missing value
df.isnull().sum()
```

```python
# Impute Missing Values using the mean or median
df['AverageTemperature'].fillna(df['AverageTemperature'].mean(), inplace=True)
df['AverageTemperatureUncertainty'].fillna(df['AverageTemperatureUncertainty'].mean(), inplace=True)
df.isnull().sum()
```

```python
# Extract the year from 'dt' column
df['Year'] = df['dt'].dt.year

# Group by Year and calculate the average temperature and uncertainty
yearly_data = df.groupby('Year')[['AverageTemperature', 'AverageTemperatureUncertainty']].mean().reset_index()

# Display the aggregated data
print(yearly_data.head())

```

```python
# Group by City and calculate the average temperature and uncertainty
city_data = df.groupby('City')[['AverageTemperature', 'AverageTemperatureUncertainty']].mean().reset_index()

# Display the aggregated data
print(city_data)
```

```python
# Group by both city and year to see how each city's temperature changes over time.
city_yearly_data = df.groupby(['City', 'Year'])[['AverageTemperature', 'AverageTemperatureUncertainty']].mean().reset_index()
print(city_yearly_data)
```

```python
# Define the bins and labels
bins = [-float('inf'), 15, 25, float('inf')]  # Define the bin ranges: Cold (< 15), Moderate (15-25), Hot (> 25)
labels = ['Cold', 'Moderate', 'Hot']  # Labels for the bins

# Create a new column 'Temperature_Category' based on the binning
df['Temperature_Category'] = pd.cut(df['AverageTemperature'], bins=bins, labels=labels)

# Display the new categorized data
print(df[['City', 'AverageTemperature', 'Temperature_Category']].head())
```

```python
# Group by city and calculate the mean of 'AverageTemperature'
top_cities = df.groupby('City')['AverageTemperature'].mean().reset_index()

# Sort the cities by 'AverageTemperature' in descending order
top_cities_sorted = top_cities.sort_values(by='AverageTemperature', ascending=False)

# Get the top 5 cities
top_5_cities = top_cities_sorted.head(5)

# Display the top 5 cities
print(top_5_cities)
```
